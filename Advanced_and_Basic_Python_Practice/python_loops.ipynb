### Loops

```python
n=0
while n<5:
    print(n)
    n+=1
```

```python
n=0
while n<5:
    print(n)
    n+=1
else:
    print("n is not less than 5 anymore")
```

```python
n=5
while n>0:
    print(n)
    n-=1
else:
    print("n is less than zero")
```

```python
n=int(input("enter the number:"))
for i in range(1,11):
    print(f"{n} X {i} = {n*i}")
```

```python
n=int(input("enter the number:"))
i=1
while i<=10:
    print(f"{n} X {i} = {n*i}")
    i+=1
```

```python
n=1
a=2
print(n,"X",a,"=",n*a)
```

```python
n=1
while n<=10:
    if n%2==0:
        print(n**2)
    else:
        print(n**3)
    n+=1
```

```python
n=1
while n<=10:
    if n%2==0:
        print(pow(n,2))
    else:
        print(pow(n,3))
    n+=1
```
