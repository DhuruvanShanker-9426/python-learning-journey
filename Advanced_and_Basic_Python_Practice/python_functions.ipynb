```python
# Instead of vowels words replacing it with whitespace
def removeVowels(s:str)->str:
    res=""
    for i in s:
        res+=i if i.lower() not in "aeiou" else " "
    return res
print(removeVowels("Dhuruvan"))
```

```python
# Removes the vowels
def removeVowels(s:str)->str:
    res=[]
    for i in s:
        if i.lower() not in "aeiou":
            res.append(i)
    return " ".join(res)
print(removeVowels("Dhuruvan"))
```

```python
name=input("Enter the string:")
left=0
right=len(name)-1
def palindrome_checker(name):
  global left
  global right
  while left<right:
    if name[left]!=name[right]:
      return False
    left+=1
    right-=1
  return True

if palindrome_checker(name):
  print(f'the word {name} is a palindrome')
else:
  print(f'the word {name} is not a palindrome')
```

```python
num1=int(input("Enter the num1:"))
num2=int(input("Enter the num2:"))
print('''1.Add
2.Sub''')
print("1.Addition\n2.Subtraction\n3.Multiplication\n4.Division\n5.Floor_Division\n6.Modulo")
operations=int(input("Enter the operation:"))
match operations:
    case 1:
        print(f'The addition of two numbers is {num1+num2}')
    case 2:
        print(f'The subtraction of two numbers is {num1-num2}')
    case 3:
        print(f'The multiplication of two numbers is {num1*num2}')
    case 4:
        print(f'The division of two numbers is {num1/num2}')
    case 5:
        print(f'The quotient of two numbers is {num1//num2}')
    case 6:
        print(f'The remainder of two numbers is {num1%num2}')
    case _:
        print("Invalid operation")
```
