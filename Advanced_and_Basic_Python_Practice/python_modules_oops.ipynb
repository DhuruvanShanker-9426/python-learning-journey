### OOPS

```python
class Person:
    def say_hi(self):
        print("Hi")
    def say_bye(self):
        print("Hello")
                
```

```python
p=Person()
```

```python
print(p)
```

```python
p.say_hi()
```

```python
p.say_bye()
```

```python
type(p)
```

```python
class Animal:
    def sound(self):
        print("Animal")
class Dog(Animal):
    def sound(self):
        print("dog")
d=Dog()
d.sound()
a=Animal()
a.sound()
```
